# Semtiment Analysis Data Processing

### Import packages

```python
import numpy as np
import pandas as pd
import spacy
import csv
import re
```

```python
nlp = spacy.load("en_core_web_lg")
```

### Load dataset
downloaded from https://www.kaggle.com/cosmos98/twitter-and-reddit-sentimental-analysis-dataset?select=Reddit_Data.csv

```python
twitter = pd.read_csv(
    "tweets_labelled.csv",
    header=0,
    delimiter=";",
    encoding = "ISO-8859-1"
)
twitter = twitter.loc[~twitter["sentiment"].isnull()]
twitter
```

```python
def char_pick(text):
    return "".join([ch for ch in text if re.match("[A-Za-z0-9 ]", ch)])
```

```python
twitter = twitter.assign(text=twitter["text"].apply(char_pick))
twitter = twitter.assign(sentiment=twitter["sentiment"].map({"positive": 1, "neutral": 0, "negative": -1}))
```

```python
twitter
```

```python
words = np.array([nlp(str(d)).vector for d in twitter["text"]])
np.savez("twitter_comment", inputs=words)
```

```python
np.save("twitter_labels", twitter["category"].to_numpy().astype(int))
```
