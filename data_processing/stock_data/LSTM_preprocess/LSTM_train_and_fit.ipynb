```python
from keras.preprocessing import sequence 
from keras.models import Sequential 
from keras.layers import Dense, Dropout, LSTM
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def input_label_generator(data, trace_back):
    X = []
    y = []
    for i in range(trace_back, len(data)):
        X.append(data[i-trace_back:i])
        y.append(data[i])
    X, y = np.array(X), np.array(y)
    X = np.reshape(X, (X.shape[0], X.shape[1], 1))
    return X, y
```

```python
stock_names = ["AAPL", "AMZN", "FB", "GOOGL", "NFLX", "TSLA"]
for stock in stock_names:
    df = pd.read_csv("data/{}_stock.csv".format(stock))
    df["Adj Close"] = df["Adj Close"] / 1000
    trace_back = 5
    year_2020 = df.loc[df["Date"]=='1/2/2018'].index[0]
    split_train = int(year_2020 * 0.8)
    split_val = year_2020
    train = df.loc[:split_train, ["Adj Close"]].to_numpy()
    val = df.loc[split_train - trace_back:split_val, ["Adj Close"]].to_numpy()
    test = df.loc[split_val - trace_back:, ["Adj Close"]].to_numpy()
    train_X, train_y = input_label_generator(train, trace_back)
    val_X, val_y = input_label_generator(val, trace_back)
    test_X, test_y = input_label_generator(test, trace_back)
    model = Sequential()
   
    model.add(LSTM(units = 200, return_sequences = True, input_shape = (train_X.shape[1], 1)))
    model.add(LSTM(units = 200, return_sequences = True))
    model.add(LSTM(units = 200))
    model.add(Dense(units = 1))
    model.compile(optimizer = 'adam', loss = 'mean_squared_error')
    model.fit(
        train_X, train_y,
        epochs = 300,
        batch_size = 64,
        validation_data = (val_X, val_y)
    )
    pred = [x[0] * 1000 for x in model.predict(test_X)]
    actual = [x[0] * 1000 for x in test_y]
    dates = df.loc[split_val:, ["Date"]]
    output = dates.assign(Pred=pred, Actual=actual)
    output.to_csv("LSTM_predictions/{}_LSTM_predictions.csv".format(stock))
```
